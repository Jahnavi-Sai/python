# Rules for a Varaiable
-> Vairable is a case sensitive i.e
Eg : age, Age, and AGE are different variables. 

-> Must start with a letter i.e 
Eg : (a-z, A-Z) or an underscore (_).

-> Can contain letters, digits (0-9), and underscores, but cannot start with a digit.

-> Cannot be a Python keyword i.e
E.g : True, False, if, for, class, etc. 

-> Vairable has no number length limit.
