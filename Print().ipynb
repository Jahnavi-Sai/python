# print format method- using {} 

Syntax: { }.format(value)

for multiple place holder Syntax : { } { } .format(value1, value2)



```
n1=8
n2=12
add=n1+n2
```

```
print('the addition of {} and {} is={}'.format(n1,n2,add))
```

```
name='chanu'
age=5
city='chennai'
```

```
print('hello guys i am {}, {} years old and i came from {}'.format(name,age,city))
```

```
r1=2.3
r2=45
r3=6.5
r4=12
r=r1+r2+r3+r4
rr=r1-r2+r3-r4
rrr=r1*r2-r3*r4-765
```

```
print(r,rr,rrr)
```

```
print('the value of r is {}, and the value of rr is {}, the value of rrr is {}'.format(r,rr,rrr))
```

```
round(rr)
```

# most short format method(f string method)

variables in curly braces and write in quotes we should start with f at starting statement

```
n=1
n1=2
add=n+n1
print(f'the addition of {n} and {n1} is= {add}')
```

```
name='chanu'
age=5
city='chennai'
print(f'my name is {name}, i am {age} yeras old ,and i came from {city}.')
```

```
r=133
r1=23
r2=213
add=r+r1+r2
avg=round((add)/3,2)
```

```
print(f' the addition of r,r1,r2 is {add}, the avg of r,r1,r2 is {avg}.')
```

```
rent=4000
water_bill=500
grocery=2000
current_bill=800
add=rent+water_bill+grocery+current_bill
```

```
print('the total expenses of the house includes rent {} water_bill {} grocery {} current_bill {} the total is {}'.format(rent,water_bill,grocery,current_bill,add))
```

```
print(f'the total expenses of the house is rent {rent}, water_bill {water_bill}, grocery {grocery}, current_bill {current_bill} and total is {add}')
```
